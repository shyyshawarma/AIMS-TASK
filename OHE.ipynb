```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Name': ['Abhinav', 'Abhijai', 'Aarushi', 'Aadarsh', 'Arnav','Aeikansh','Hemang','Hemant'],
    'Age':[24, 21, 22, 27, 29,12,32,54],
    'Gender' : ["M", "M", "F","M","M","M","M" ,"M"],
    'Score': [85.0, 99, 99, 88.0, 76.5,34,79,90],
    'Highmarks':[99,99,94,90,97,76,80,93],
    'PASS/FAIL' : ["p", "p", "p","p","p","f","p" ,"p"]
}




```

```python
d=pd.DataFrame(data)
```

```python
d
```

```python
features=d.columns
    
```

```python
features
```

```python
s = (d.dtypes == 'object')
object_cols = list(s[s].index)

object_cols=[col for col in features
           if d[col].dtype=='object']
cols_with_unique = [col for col in object_cols
                    if d[col].nunique()<3]

    
```

```python
object_cols
```

```python
cols_with_unique
```

```python
d
```

```python
def myoe(column):
    unique=list(set(column));
    return {val:[1 if v==val else 0 for v in column] for val in unique}
```

```python
for i in cols_with_unique:
    new_cols_dict=myoe(d[i]);
    new_d=pd.DataFrame(new_cols_dict)
    d=d.join(new_d)
```
